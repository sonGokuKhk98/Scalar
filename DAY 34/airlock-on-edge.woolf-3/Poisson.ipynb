```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

### P[X = 0]

```python
np.exp(-3)
```

### P[X = 1]

```python
3 * np.exp(-3)
```

```python
poisson.pmf(mu = 3, k = 1)
```

## P[X <= 1] 

```python
## P[X <= 1] : prob. of observing less than or exual to 1 accident next day

P[X <= 1]  = P[X = 0] + P[X = 1]  
```

```python
# P[X = 0] + P[X = 1]=
poisson.pmf(mu = 3, k = 0) + poisson.pmf(mu = 3, k = 1)
```

```python
# P[X <= 1] = P[X = 0] + P[X = 1] = poisson.cdf(mu = 3, k = 1)
poisson.cdf(mu = 3, k = 1)
```

```python
# P[X <=7] = 
poisson.cdf(mu = 3, k = 7)
```

```python
X_values = np.arange(0,11)
X_values

y_values = []

for k in X_values:
    y = poisson.pmf(mu = 3, k = k)
    y_values.append(np.round(y,4))

# y_values

sns.barplot(X_values, y_values)
plt.title("lambda = 3 acc/day : Average Rate")
plt.show()
```
