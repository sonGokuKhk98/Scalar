## More on missing values

We have learnt how to fill values of one column using an estimator

#### Let's play around a bit more

We will look at two new methods, ```ffill``` and ```bfill```

````python
sample_df = data.head(10)
sample_df
````

````python
sample_df.fillna(method='ffill')
````

#### What do we notice here?

Forward fill basically replaces the null value with the value just before it.

Thus, at row 2, the NaN value of `1:30:00` column gets replaced by 12 value of row 1

````python
sample_df.fillna(method='bfill')
````

#### Now what did bfill do exactly?

Opposite to forward fill, it replaced the NaN values with the values after them.

This replacement works upwards, while forward fill works downward.

#### Now let's look at another interesting concept, the change of dtype by fillna

````python
sample_df.info()
````

As we can see, the dtype of `1:30:00` column is float64

Now what happens if we use fillna on this?

````python
sample_df.fillna(0).info()
````

It's the same, but now, what if we use a string in fillna?

````python
sample_df.fillna("Couldn't Take Reading").info()
````

#### When we replace the value with string, it changes the dtype of object.

Remember from our earlier discussions, a string based column is considered an object dtype

This is what exactly happened here.

### Pivot_table



#### Let's first use melt and pivot to restructure our data as we did in the lecture

````python
data_melt = pd.melt(data,id_vars = ['Date', 'Drug_Name', 'Parameter'], 
            var_name = "time",      
            value_name = 'reading')  
data_tidy = data_melt.pivot(index=['Date','time', 'Drug_Name'], 
                                        columns = 'Parameter',  
                                        values='reading') 
data_tidy = data_tidy.reset_index()
data_tidy.columns.name = 'None'
data_tidy.head()
````

Now suppose we want to find some insights, like **mean temperature day wise**

#### Can we use pivot to find the day-wise mean value of temperature for each drug?

````python
data_tidy.pivot(index=['Drug_Name'],  
                columns = 'Date',
                values=['Temperature']) 
````

#### Why did we get an error?

- We need to find the **average** of temperature values throughout a day

- If you notice, the error shows **duplicate entries.**

Hence the index values should be unique entry for each row.

#### What can we do to get our required mean values then?


````python
pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Temperature'], aggfunc=np.mean)
````

This function is similar to pivot, with an extra feature of an aggregator

#### How does `pivot_table` work?
- The initial parameters are same as how we do in `pivot()`
- As an extra parameter, we pass the **type of aggregator**

Note:

- We could have done this using `groupby` too

- In fact, `pivot_table` uses `groupby` in the backend to group the data and perform the aggregration

- The only difference is in the type of output we get using both functions



#### Similarly, what if we want to find the minimum values of temperature and pressure on a particular date?

````python
pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Temperature', 'Pressure'], aggfunc=np.min)
````
