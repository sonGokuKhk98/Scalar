```python
df_tidy=pd.pivot(df_melt,
         index=["Date","Drug_Name","Time"],
         columns="Parameter",
         values="Result").reset_index()
df_tidy.columns.name=None
```

```python
def temp_mean(x):
    x["Temp_avg"]=np.round(x["Temperature"].mean(),2)
    return x
```

```python
df_tidy = df_tidy.groupby("Drug_Name").apply(temp_mean)
```

```python
df_tidy["Temperature"].fillna(df_tidy["Temp_avg"],inplace=True)
df_tidy
```

```python
df_tidy.isna().sum()
```

```python
def pres_mean(x):
    x["Pres_avg"]=np.round(x["Pressure"].mean(),2)
    return x
```

```python
df_tidy=df_tidy.groupby("Drug_Name").apply(pres_mean)

df_tidy["Pressure"].fillna(df_tidy["Pres_avg"],inplace=True)
df_tidy
```

```python
temp_points=[7,20,24,27,58]
temp_labels=["low","medium","high","very_high"]
```

```python
df_tidy["Temp_cat"]=pd.cut(df_tidy["Temperature"],
                          bins=temp_points,
                          labels=temp_labels)
df_tidy
```
