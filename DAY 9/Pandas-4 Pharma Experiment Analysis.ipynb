```python
# Deal with missing values
import pandas as pd 
df = pd.read_csv("Pfizer_1.csv")
df # Wide format is having same type of columns haing a greater count
```

```python
df.info() #3 Types of data
```

```python
# Melting
# We can have a single column as "time"
# We need to melt it to the long format from wide format
pd.melt(df,id_vars=["Date","Drug_Name","Parameter"])
# AS we can see 15 columns converted into 5 columns
```

```python
pd.melt(df,id_vars=["Date","Drug_Name","Parameter"],var_name = "Time",value_name="Result")
```

```python
# Lets say we took away "Parameter" column
pd.melt(df,id_vars=["Date","Drug_Name"],var_name = "Time",value_name="Result")
# The parameter is repeating
# We don't get any relevant data, we first iterate through the parameter column values and then come to the result
```

```python
df_melt = pd.melt(df,id_vars=["Date","Drug_Name","Parameter"],var_name = "Time",value_name="Result")
#Lets see the difference in number of rows and columns
print(df.shape)
df_melt.shape
```

```python
# To bring it back to wide format from long format

#WAY 1
df_pivot=df_melt.pivot(index=["Date","Drug_Name","Parameter"],
         columns="Time",
         values="Result")

df_pivot
```

```python
#WAY 2
pd.pivot(df_melt, index=["Date","Drug_Name","Parameter"],
         columns="Time",
         values="Result")
# We still need to bring it to the each row values format
```

```python
df_pivot.reset_index() #Here the time column is in the index but that isn't time
# Messed up the data formating
```

```python
df_pivot.columns
```

```python
df_pivot.columns.name = None
df_pivot.reset_index()
```

```python
df_melt  # Here we see each of the drug and date is repeating twice 
```

```python
df_tidy = pd.pivot(df_melt, index=["Date","Drug_Name","Time"],
                   columns="Parameter",
                   values="Result").reset_index()
df_tidy
```

```python
df_tidy.columns
```

```python
df_tidy.columns.name = None
df_tidy #Much more interpretiable than before
```

```python
# Pivot table
mean =pd.pivot_table(df_tidy,index="Drug_Name",
               columns="Date",
               values=["Pressure","Temperature"],
               aggfunc="mean",
               margins=True).reset_index()
mean
# ALL means row wise grand total and column wise grand totals
```

```python
# Pivot table
mean =pd.pivot_table(df_tidy,index="Drug_Name",
               columns="Date",
               values=["Pressure","Temperature"],
               aggfunc=["min","max"],
               margins=True).reset_index()
mean
```
