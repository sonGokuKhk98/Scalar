```python
#Pivot_table

#For each drug, give me the mean temp on all days
```

```python
pd.pivot_table(df_tidy,
         index="Drug_Name",
         columns="Date",
         values=["Pressure","Temperature"],
         aggfunc="mean" 
              )
```

```python
pd.pivot_table(df_tidy,
         index="Drug_Name",
         columns="Date",
         values=["Pressure","Temperature"],
         aggfunc="min",
         margins = True
              )
```

```python
pd.pivot_table(df_tidy,
         index="Drug_Name",
         columns="Date",
         values=["Pressure","Temperature"],
         aggfunc=["min","max"],
         margins = True
              )
```
