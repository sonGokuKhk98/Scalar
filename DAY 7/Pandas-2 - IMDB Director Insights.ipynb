```python
data = pd.merge(movies,directors,left_on="director_id",right_on="id",how="inner")
data
```

```python
data.drop(columns=["director_id","id_y"],inplace=True)
```

```python
data.info() #Missing values in gender
```

```python
data.describe(include="all").T # .T Gives a better view
```

```python
data["budget"] = (data["budget"]/1000000).round(2)
data
```

```python
data["vote_average"]>7 #Mask
```

```python
data[data["vote_average"]>7]
```

```python
data.loc[data['vote_average']>7] 
```

```python
#For all those movies give me director name and title
data.loc[data["vote_average"]>7,["title","director_name"]]
```

```python
#For all those movies give me year greater than 2015
data.loc[(data["vote_average"]>7) & (data["year"]>=2015)] 
```

```python
# Movies released on a Friday or Saturday
data.loc[(data["day"]=="Friday") | (data["day"]=="Saturday")] 
```

```python
data.sort_values(by=["popularity"],ascending=False).head()
```

```python
data.loc[data["director_name"]=="Christopher Nolan"]
```

```python
data.loc[data["director_name"]=="Christopher Nolan",["title","year"]]
```

```python
# We can create our own function and vectorize it
def myfunc(data):
    if data == 'Male':
        return 1
    elif data == 'Female':
        return 0
    else:
        return np.nan
```

```python
data['gender'] = data["gender"].apply(myfunc)
data
```

```python
# Simple method to create new columns
data["profit_simple_method"] = data['revenue'] - data['budget']
data
```

```python
def profit_func(x):
    return x["revenue"] - x["budget"]
```

```python
data["profit_apply_method"]= data[["revenue","budget"]].apply(profit_func,axis=1) #Row wise
```
