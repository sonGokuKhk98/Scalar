```python
import pandas as pd
movies = pd.read_csv("movies.csv")
directors = pd.read_csv("directors.csv")

movies
```

```python
directors
```

```python
data = pd.merge(movies,directors,left_on="director_id",right_on="id",how="inner")
data
```

```python
data.drop(columns=["Unnamed: 0_x","Unnamed: 0_y"],inplace=True)
```

```python
#Convert a Series to a dataframe
data["director_name"].value_counts().to_frame()
```

```python
# To count all the movies by christopher nolan
data.loc[data["director_name"]=="Christopher Nolan"]["title"].count()
```

```python
data.groupby("director_name").ngroups #Number of unique groups
```

```python
data.groupby("director_name").groups #Number of unique groups gives indice
# Follows explicit indexing
```

```python
data.groupby("director_name").get_group("Christopher Nolan")
```

```python
data.groupby("director_name")["title"].count()
```

```python
data.groupby("director_name")["year"].agg(["min","max"])
```

```python
import numpy as np
data.groupby("director_name")["year"].agg([np.min,np.max])
```

```python
# Want to filter out high budgeted director
data.loc[data['budget']>100000000]["director_name"] # Greater than 100M
```

```python

dir_high_budget= data.groupby("director_name")["budget"].max().reset_index()
dir_high_budget
```

```python
dir_high_budget.loc[dir_high_budget["budget"]>=100000000]
```

```python
names = dir_high_budget.loc[dir_high_budget["budget"]>=100000000,"director_name"]
names
```

```python
data.loc[data["director_name"].isin(names)] # We can get all the details 
```

```python
# More optimal method

data.groupby("director_name").filter(lambda x: x["budget"].max()>=100000000)
```

```python
# Get risky movies
def func(x):
    x['risky'] = x['budget'] - x['revenue'].mean() >= 0 #Risky
    return x

data_risky = data.groupby("director_name").apply(func)
```

```python
data_risky.loc[data_risky["risky"]] # 131 movies are risky
```
