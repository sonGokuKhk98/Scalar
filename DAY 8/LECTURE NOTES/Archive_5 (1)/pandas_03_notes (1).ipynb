```python
import pandas as pd
import numpy as np
!gdown 1s2TkjSpzNc4SyxqRrQleZyDIHlc7bxnd
!gdown 1Ws-_s1fHZ9nHfGLVUQurbHDvStePlEJm
movies = pd.read_csv('movies.csv', index_col=0)
directors = pd.read_csv('directors.csv',index_col=0)
data = movies.merge(directors, how='left', left_on='director_id',right_on='id')  
data.drop(['director_id','id_y'],axis=1,inplace=True)
```

```python
data["budget"]=(data["budget"]/1000000).round(2)
data["revenue"]=(data["revenue"]/1000000).round(2)
```

```python
data.groupby("director_name").filter(lambda x:x["budget"].max()>=100)
```

```python
def func(x):
    x['risky']= x['budget']-x['revenue'].mean()>=0 #risky
    return x
```

```python
data.groupby("director_name").apply(func)
```

```python
data_risky = data.groupby("director_name").apply(func)
data_risky
```

```python
data_risky.loc[data_risky["risky"]==False]  
```

```python
data_risky["risky"]
```

```python
data_multi = data.groupby("director_name")[["year","title"]].agg({"year":["min","max"],
                                                     "title":["count"]})
data_multi
```

```python
data_multi.columns
```

```python
data_multi.columns = ["_".join(col) for col in data_multi.columns] 
```

```python
data_multi.columns
```

```python
data_multi 
```

```python
data_multi.reset_index() 
```

```python
data.groupby("director_name")[["year","title"]].agg(year_min=("year","min"),
                                                   year_max=("year","max"),
                                                   title_count=("title","count"),)
```

```python
data_multi.reset_index(inplace=True)
```

```python
data_multi
```

```python
data_multi["years_active"]=data_multi["year_max"]-data_multi["year_min"]
data_multi
```

```python
data_multi["movies_per_year"]=data_multi["title_count"]/data_multi["years_active"]
data_multi
```

```python
data_multi.sort_values(by="movies_per_year",ascending=False)
```
