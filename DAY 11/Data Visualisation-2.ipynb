```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Since it has two variables, this is a bivariate analysis | Two features are involved.
x = [0,1,2]
y = [5,6,9]

plt.plot(x,y)
```

```python
data  = pd.read_csv('final_vg.csv')
```

```python
# Univariate analysis : Categorical data
data["Genre"] 
# We can see if we can get count of each of the data which is analysis on a single feature
cat_counts = data["Genre"].value_counts()
print(cat_counts)
```

```python
# We can use a bar chat to get frequency of occurances.
# Histogram is different, we create bins with continous data , and then we see what data points fall in which bin
x = cat_counts.index
y = cat_counts
plt.bar(x,y)
```

```python
plt.figure(figsize=(12,8)) #6.4 inch by 4.8 inch is default
# To make the tics prominent 
x = cat_counts.index
y = cat_counts
plt.bar(x,y)
```

```python
plt.figure(figsize=(12,8)) #6.4 inch by 4.8 inch is default
# To make the tics prominent 
x = cat_counts.index
y = cat_counts
plt.bar(x,y)
plt.xticks(rotation=90) #The output shows what it does internally
plt.show()
```

```python
x = cat_counts.index
y = cat_counts
plt.bar(x,y,width=0.2,color = "maroon")
plt.xticks(rotation=90, fontsize=12, ) #The output shows what it does internally
plt.show()
```

```python
x = cat_counts.index
y = cat_counts
plt.bar(x,y,width=0.2,color = "maroon")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=90, fontsize=12, ) #The output shows what it does internally
plt.show()
```

```python
x = cat_counts.index
y = cat_counts
plt.bar(x,y,width=0.2,color = "maroon")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Games Per Genre")
plt.xticks(rotation=90, fontsize=12, ) #The output shows what it does internally
plt.show()
```

```python
x = cat_counts.index
y = cat_counts
plt.figure(figsize=(12,8))
plt.bar(x,y,width=0.2,color = "maroon")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Games Per Genre")
plt.xticks(rotation=90, fontsize=12, ) #The output shows what it does internally
#plt.show()


plt.figure(figsize=(12,8))
x = cat_counts.index
y = cat_counts
plt.bar(x,y,width=0.2,color = "orange")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Games Per Genre")
plt.xticks(rotation=90, fontsize=12 ) #The output shows what it does internally
plt.show()
```
