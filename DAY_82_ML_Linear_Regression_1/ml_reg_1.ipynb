```python
import pandas as pd
```

```python
df = pd.read_csv('cars24-car-price-cleaned.csv')
```

* We want to convert the categorical data like make, model into categorical data

```python
df['make'].unique()
```

```python
df.groupby('make')['selling_price'].mean().to_dict()
```

* As we see LAMBORGHINI, LEXUS have the same means (20.902500, 20.902500)
* We can iterate on values that have same means and change them by precision, so that the make has it's own representation

```python
#  grouping by make and taking mean of it's Selling price, to replace each value with mean of the group
df['make'] = df.groupby('make')['selling_price'].transform('mean')
df['model'] = df.groupby('model')['selling_price'].transform('mean')
```

* The 'make' and 'model' has been coverted into means (numerical) of it's make group selling price
