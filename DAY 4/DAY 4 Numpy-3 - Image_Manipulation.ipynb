```python
y = np.vectorize(math.log) #Vecorized the function in numpy
y(a) #Vectorise 
```

```python
def custom_function(x):
    if x%2 == 0:
        x+=2
    else:
        x+=3
    return x
y1 = np.vectorize(custom_function) # Can be directly implemented as a function over all elements in a numpy array
y1(a)
```

Broadcasting

A (3,2)   B(4,2)

* STEP 1 : Dimension of A and B match ?
         Dim(A) = Dim(B) = 2
* STEP 2 : Check if each of the dimensions match?
3!=4 but 2=2

A(2,)    B(3,2)

* STEP 1 : Dimension of A and B match ?
         No -> Dim(A) < Dim(B)
                1 < 2
                Make A(1,2) and B(3,2)
* STEP 2: 2 = 2 yes
        1 = 3 No
        We can scale it up 3 = 3
