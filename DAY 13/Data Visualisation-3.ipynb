```python
top_3_pub = data["Publisher"].value_counts().index[:3]
top_3_gen = data["Genre"].value_counts().index[:3]
top_3_plat = data["Platform"].value_counts().index[:3]
```

```python
top3_data = data.loc[data["Publisher"].isin(top_3_pub) & data["Genre"].isin(top_3_gen) & data["Platform"].isin(top_3_plat)]
top3_data
```

```python
sns.scatterplot(data= top3_data, x="NA_Sales",y="Global_Sales")
```

```python
sns.scatterplot(data= top3_data, x="JP_Sales",y="Global_Sales")
```

```python
sns.scatterplot(data= top3_data, x="EU_Sales",y="Global_Sales",hue="Genre")
```

```python
sns.lineplot(data=top3_data,
             x="Year",
             y="Global_Sales",
             hue="Genre")
#The shaded part is confidence interval
```

```python
sns.lineplot(data=top3_data,
             x="Year",
             y="Global_Sales",
             hue="Genre",
             estimator=)
#The shaded part is confidence interval
```

```python
# 2 methodscreate subplots

fig= plt.figure(figsize=(12,10))
#Two rows and 3 columns 1 is the position
plt.subplot(2,3,1)
sns.scatterplot(data= top3_data,x="JP_Sales",y="Global_Sales",color ="red")
plt.title("JP_Sales vs Global_Sales")

plt.subplot(2,3,3)
sns.scatterplot(data= top3_data,x="EU_Sales",y="Global_Sales",color ="blue")
plt.title("EU_Sales vs Global_Sales")

plt.subplot(2,3,5)
sns.scatterplot(data= top3_data,x="NA_Sales",y="Global_Sales",color ="red")
plt.title("NA_Sales vs Global_Sales")

fig.suptitle("Subplots")
plt.show()
```

```python
fig,ax = plt.subplots(2,3,figure=(12,10))
#0th row 1st column
ax[0,1].scatter(data=top3_data,x="JP_Sales",y="Global_Sales",color="red")

ax[1,0].scatter(data=top3_data,x="EU_Sales",y="Global_Sales",color="blue")
ax[1,1].scatter(data=top3_data,x="NA_Sales",y="Global_Sales",color="orange")

ax[1,2].scatter(data=top3_data,x="NA_Sales",y="Global_Sales",color="orange")
plt.show()

```
